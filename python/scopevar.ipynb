```python
# local variable
def book():
    sub="maths"
    print('my book is :',sub)
book()    
```

```python
# enclosed 
def flavor():
    choose='venilla'
    def checkout(): 
        print('your cake flavor:',choose) 
    checkout()
flavor()    
```

```python
# global variable
user='ramesh'
def my_home():
    print(user,'welcome to my home')
def my_office():
    print(user,'welcome to my office')    
my_home()
my_office()    
```

```python
subject='maths'
def student():
    name='ramesh'
    def checkout():
        marks=33
        print(f'{name} sub{subject} {marks}')
    checkout()
student()    
```
