type caseting

```python
#typecasting------>astype()
import numpy as np
arr=np.array([1,2,3,4])
arr
```

```python
arr3=arr.astype(np.float64)
arr3
```

```python
arr4=arr3.astype(np.int64)
arr4.dtype
```

```python
# type casting errors
arr=np.array(["rame","33",'5.5'])
arr1=arr.astype(np.int64)
arr1
```

```python
arr=np.array([1,1.2,4])
arr1=arr.astype(np.float64)
arr1
```
