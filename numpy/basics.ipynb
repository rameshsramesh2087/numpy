```python
import numpy as np
print(np.__version__)
```

```python
arr=np.array([10.1,12,13,15],dtype=int)
print(arr)
print([1,2,3])
print('shape of id:',arr.shape)
print(arr.size)
print(arr.ndim)
print(arr.dtype)
```

```python
arr2d=np.array([[1,2,3,4],[5,6,7,8]],dtype=float)
print(arr2d)
print(type(arr2d))
print('shape of 2d:',arr2d.shape)
print(arr2d.size)
print(arr2d.ndim)
print(arr2d.dtype)
```

```python
print(np.ones((5,5),dtype=int))
print(np.zeros((4,4),dtype=int))
```

```python
print(np.full((2,3),100))
```
